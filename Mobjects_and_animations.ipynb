The mathematical objects that are displayed in Manim is called 'Mobjects'.

Mobject is the parent class that has many descendent classes 

such as VMojbect, ImageMobject, PMobject ,ValueTracker, Group/VGroup, etc.



VMobject

VMobject is essentially 'Bezier curve'; represent the complex curves using polynomials. 


ImageMobject is essentially matrix pixels (Later in Chapter 6).
ImageMobject is defined by Pixel array whereas VMobject is defined by Bezeir curve.

Group/VGroup

Groups are containers for Mobjects. They are used in general for arranging Mobjects for styles (Chapter 7). 

Group can contain any type of Mobject.

VGroup can only contain VMobjects or subclasses of VMobject. 

Group can contain any Mojbect, even VMobjects (like Circle), so there will be no error here.



```python
%%manim -qm -v WARNING GroupvsVGroup

from manim import *

class GroupvsVGroup(Scene):
    def construct(self):
        image = ImageMobject(np.uint8([list(range(255))]))
        circle = Circle()
        grp = Group(image, circle)
        self.add(grp)
```
